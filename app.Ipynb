```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import time 
import string
import re
```

```python
tweet_df = pd.read_csv('tweets.csv')
print(tweet_df.head(10))
```

```python
# Remove all the urls in the text.

def remove_urls(text):
  url = re.sub('http://\S+|https://\S+', "", text)
  return url

# Remove all the Punctuation in the text.
def remove_punctuation(text):
  punct = text.translate(str.maketrans("", "", string.punctuation))
  return punct

# Remove all the Emoji in the text.
import emoji

def remove_emojis(text):
  return emoji.replace_emoji(text, replace="")
```

```python
tweet_df['text'] = tweet_df.text.map(remove_urls)
tweet_df['text'] = tweet_df.text.map(remove_punctuation)
tweet_df['text'] = tweet_df.text.map(remove_emojis)
print(tweet_df['text'].head(50))
```

```python
# Remove all the stop words

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Stop Words: A stop word is a commonly used word (such as “the”, “a”, “an”, “in”) that a search engine
# has been programmed to ignore, both when indexing entries for searching and when retrieving them
# as the result of a search query.

stop_word = set(stopwords.words('english'))

def remove_stopword(text):
  filter_words = [word.lower() for word in text.split() if word.lower() not in stop_word]
  return ' '.join(filter_words)
```

```python
print(stop_word)
```

```python
tweet_df['text'] = tweet_df.text.map(remove_stopword)
print(tweet_df.text )
```

```python
# Count unique words
from collections import Counter

def count_word(text_cal):
  count = Counter()
  for text in text_cal.values:
    for word in text.split():
      count[word] += 1
  return count

counter = count_word(tweet_df.text)
num_unique_words = len(counter)

print(len(counter))
print(counter)
print(counter.most_common(5))
```

```python
# Split data into trainning and validation set

train_size = int(tweet_df.shape[0] * 0.8)
print(train_size)

train_df = tweet_df[:train_size]
val_df = tweet_df[train_size:]

print(train_df)
print('#########################')
print(val_df)
```

```python
# split text and lables

train_sentences = train_df.text.to_numpy()
train_lables = train_df.target.to_numpy()

val_sentences = val_df.text.to_numpy()
val_lables = val_df.target.to_numpy()

print(train_sentences.shape)
print(val_sentences.shape)
```

```python
# Tokenize
from tensorflow.keras.preprocessing.text import Tokenizer

# vectorize a text corpus by turning each text into a sequence of integers
tokenizer = Tokenizer(num_words = num_unique_words)
tokenizer.fit_on_texts(train_sentences)

# each word has unique index
word_index = tokenizer.word_index

word_index
```

```python
train_sequences = tokenizer.texts_to_sequences(train_sentences)
val_sequences = tokenizer.texts_to_sequences(val_sentences)

print(train_sentences[10:15])
print(train_sequences[10:15])
```

```python
# Pad the sequences to have the same length
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Max number of words in a sequence
max_length = 20

train_padded = pad_sequences(train_sequences, maxlen= max_length, padding='post', truncating='post')
val_padded = pad_sequences(val_sequences, maxlen = max_length, padding='post', truncating='post')

train_padded.shape, val_padded.shape

train_padded[10]
```

```python
print(train_sentences[10])
print(train_sequences[10])
print(train_padded[10])
```

```python
# Check reversing the indices

# flip (key, value)
revers_word_index = dict([(idx, word) for (idx, word) in word_index.items()])
revers_word_index
```

```python
def decode(sequence):
  return " ".join([revers_word_index.get(idx, '?') for idx in sequence])
```

```python
decoded_text = decode(train_sequences[10])

print(train_sequences[10])
print(decoded_text)
```

```python
# Create LSTM model
from tensorflow.keras import layers, regularizers

# Embedding: https://www.tensorflow.org/tutorials/text/word_embeddings
# Turns positive integers (indexes) into dense vectors of fixed size. (other approach could be one-hot-encoding)

# Word embeddings give us a way to use an efficient, dense representation in which similar words have
# a similar encoding. Importantly, you do not have to specify this encoding by hand. An embedding is a
# dense vector of floating point values (the length of the vector is a parameter you specify).

model = keras.models.Sequential()
model.add(layers.Embedding(num_unique_words,
                           32,
                           input_length=max_length))

# The layer will take as input an integer matrix of size (batch, input_length),
# and the largest integer (i.e. word index) in the input should be no larger than num_words (vocabulary size).
# Now model.output_shape is (None, input_length, 32), where `None` is the batch dimension.


model.add(layers.LSTM(64,
                      dropout=0.1,
                      kernel_regularizer = regularizers.l2(0.5)))

model.add(layers.Dense(1, activation="sigmoid",
          kernel_regularizer = regularizers.l2(0.5)))

model.summary()
```

```python
loss = keras.losses.BinaryCrossentropy(from_logits=False)
optim = keras.optimizers.Adam(learning_rate = 0.001)
metrics = ['accuracy']

model.compile(loss=loss, optimizer=optim, metrics=metrics)
```

```python
model.fit(train_padded, train_lables, epochs=100, validation_data=(val_padded, val_lables), verbose=2)
```

```python
predictions = model.predict(train_padded)
predictions = [1 if p > 0.5 else 0 for p in predictions]
```
